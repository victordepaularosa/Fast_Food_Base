```
caminho_origem = "/FileStore/tables/projeto_fast_food/silver/"
caminho_destino = "/FileStore/tables/projeto_fast_food/gold/"
```

```
tabela_origem = "silver_fast_food"

df_fast_food = spark.read.format("delta").load(caminho_origem + tabela_origem)

df_fast_food.createOrReplaceTempView("vw_fast_food")
```

```
tabela_origem = "silver_restaurantes"

df_restaurantes = spark.read.format("delta").load(caminho_origem + tabela_origem)

df_restaurantes.createOrReplaceTempView("vw_restaurantes")
```

```
tabela_origem= "silver_pratos"

df_pratos = spark.read.format("delta").load(caminho_origem + tabela_origem)

df_pratos.createOrReplaceTempView("vw_pratos")
```

```
tabela_destino= "gold_d_pratos"

df_pratos.write.mode("overwrite").parquet(caminho_destino + tabela_destino)
```

```
tabela_destino= "gold_d_restaurantes"

df_restaurantes.write.mode("overwrite").parquet(caminho_destino + tabela_destino)
```

```
df_fast_food = spark.sql(f"""
    SELECT 
        rest.ID_RESTAURANTES       AS KEY_RESTAURANTES,
        prat.ID_PRATOS             AS KEY_ID_PRATOS,
        ff.CALORIAS,
        ff.CALORIAS_DE_GORDURA,
        ff.CALORIAS_TOTAIS,
        ff.GORDURA_SATURADA,
        ff.GORDURA_TRANS,
        ff.COLESTEROL,
        ff.SODIO,
        ff.TOTAL_DE_CARBOIDRATOS,
        ff.FIBRAS,
        ff.ACUCAR,
        ff.PROTEINAS,
        ff.VITAMINA_A,
        ff.VITAMINA_C,
        ff.CALCIO,
        ff.SALADA
    
    FROM vw_fast_food AS ff

    LEFT JOIN vw_restaurantes AS rest
    ON ff.RESTAURANTES = rest.RESTAURANTES

    LEFT JOIN vw_pratos AS prat
    ON ff.PRATOS = prat.PRATOS
""")
```

```
tabela_destino= "gold_f_fast_food"

df_fast_food.write.mode("overwrite").parquet(caminho_destino + tabela_destino)
```
